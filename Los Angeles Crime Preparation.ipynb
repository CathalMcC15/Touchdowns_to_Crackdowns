### Sorting Date and Time

```python
la['Date Occurred'] = la['Date Occurred'].str.split(' ').str[0]
```

```python
la.head()
```

```python
la['Time Occurred'] = la['Time Occurred'].apply(str)
```

```python
la['Date'] = la['Date Occurred'] + ' ' + la['Time Occurred']
```

```python
la.head()
```

```python
def totime(date):
    date, time = date.split()
    if len(time) <= 2:
        time = "00" + time
    
    return pd.to_datetime(' '.join([date, time]), format="%Y-%m-%d %H%M")
```

```python
totime('2013-02-09 5')
```

```python
la['Date'] = la['Date'].apply(lambda x: totime(x))
```

```python
la = la.set_index('Date')
```

### Sorting Crimes

```python
t1 = la[la['Crime Code Description'] == "BATTERY - SIMPLE ASSAULT"]
t2 = la[la['Crime Code Description'] == "INTIMATE PARTNER - SIMPLE ASSAULT"]
t3 = la[la['Crime Code Description'] == "VANDALISM - FELONY ($400 & OVER, ALL CHURCH VANDALISMS) 0114"]
t4 = la[la['Crime Code Description'] == "VANDALISM - MISDEAMEANOR ($399 OR UNDER)"]
t5 = la[la['Crime Code Description'] == "ASSAULT WITH DEADLY WEAPON, AGGRAVATED ASSAULT"]
t6 = la[la['Crime Code Description'] == "INTIMATE PARTNER - AGGRAVATED ASSAULT"]
t7 = la[la['Crime Code Description'] == "BATTERY POLICE (SIMPLE)"]
t8 = la[la['Crime Code Description'] == "ARSON"]
t9 = la[la['Crime Code Description'] == "THROWING OBJECT AT MOVING VEHICLE"]
t10 = la[la['Crime Code Description'] == "ASSAULT WITH DEADLY WEAPON ON POLICE OFFICER"]
t11 = la[la['Crime Code Description'] == "CHILD ABUSE (PHYSICAL) - AGGRAVATED ASSAULT"]
t12 = la[la['Crime Code Description'] == "OTHER ASSAULT"]
```

```python
frames = [t1, t2, t3, t4, t5, t6, t7, t8, t9, t10, t11, t12]
la = pd.concat(frames)

frames1 = [t1, t5, t7, t10, t12]
assault = pd.concat(frames1)

frames2 = [t2, t6, t11]
domestic = pd.concat(frames2)

frames3 = [t3, t4, t8, t9]
vandalism = pd.concat(frames3)
```
