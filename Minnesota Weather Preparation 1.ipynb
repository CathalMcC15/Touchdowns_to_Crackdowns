Find the average between 280.690 and 278.850. This is equal to 279.770.

```python
df['Temperature'] = df['Temperature'].fillna(279.770)
```

```python
print(df[df["Temperature"].isnull()])
```

We want to convert the temperature to degrees Celcius from kelvin. The formula to do this is T(°C)=T(K)-273.15, Therefore, we create a new column called 'Celcius' and use this formula. 

```python
df["Celcius"] = df["Temperature"] - 273.15
```

```python
df.head()
```

```python
df = df.drop(df.columns[[0]], axis=1)
```

### Weather Grading System

```python
minWea = df
```

```python
minWea.head()
```

```python
pt = minWea.groupby('Description')[('Celcius')].count()
pt.sort_values(ascending = False)
```

```python
conditions = [
    (minWea['Description'] == 'sky is clear') | (minWea['Description'] == 'few clouds'),
    (minWea['Description'] == 'overcast clouds') | (minWea['Description'] == 'broken clouds') 
    | (minWea['Description'] == 'scattered clouds') | (minWea['Description'] == 'fog') | (minWea['Description'] == 'haze')]
choices = ['3', '2']
minWea['DescRate'] = np.select(conditions, choices, default='0')
print(minWea)
```

```python
minWea['Celcius'].mean()
```

```python
conditions = [
    (minWea['Celcius'] <= 7),
    (minWea['Celcius'] >= 15)]
choices = ['0', '3']
minWea['TempRate'] = np.select(conditions, choices, default='2')
print(minWea)
```

```python
minWea.head()
```

```python
minWea['Wind Speed'].mean()
```

```python
conditions = [
    (minWea['Wind Speed'] <= 3),
    (minWea['Wind Speed'] > 3) | (minWea['Wind Speed'] < 6) ]
choices = ['3', '2']
minWea['WindRate'] = np.select(conditions, choices, default='1')
print(minWea)
```

```python
minWea.head()
```

```python
minWea.info()
```

```python
minWea['DescRate'] = minWea['DescRate'].astype(int)
```

```python
minWea['TempRate'] = minWea['TempRate'].astype(int)
```

```python
minWea['WindRate'] = minWea['WindRate'].astype(int)
```

```python
minWea.info()
```

```python
minWea['Overall'] = minWea["DescRate"] + minWea["TempRate"] + minWea["WindRate"]
```

```python
minWea.head()
```

```python
conditions = [
    (minWea['Overall'] >= 8), 
    (minWea['Overall'] >= 6) & (minWea['Overall'] < 8) ]
choices = ['Good', 'Moderate']
minWea['Weather'] = np.select(conditions, choices, default='Bad')
```
