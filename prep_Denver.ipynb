```python
import os
import sys
import hashlib
import pandas as pd
import numpy as np

if not os.path.exists( "../../data/raw/Den_crime.csv" ):
    print("Missing dataset file")
    
crime = pd.read_csv("../../data/raw/Den_crime.csv", index_col = "FIRST_OCCURRENCE_DATE",
                    parse_dates = True)
```

```python
crime.head()
```

```python
col_list = ['OFFENSE_TYPE_ID', 'GEO_LON', 'GEO_LAT']
```
