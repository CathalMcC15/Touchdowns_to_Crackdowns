We want to convert the temperature to degrees Celcius from kelvin. The formula to do this is T(Â°C)=T(K)-273.15, Therefore, we create a new column called 'Celcius' and use this formula. 

```python
df["Celcius"] = df["Temperature"] - 273.15
```

```python
df.head()
```

```python
df = df.drop(df.columns[[0]], axis=1)
```

### Weather Grading System

```python
denWea = df
```

```python
denWea.head()
```

```python
pt = denWea.groupby('Description')[('Celcius')].count()
pt.sort_values(ascending = False)
```

```python
conditions = [
    (denWea['Description'] == 'sky is clear') | (denWea['Description'] == 'few clouds') | (denWea['Description'] == 'scattered clouds'),
    (denWea['Description'] == 'overcast clouds') | (denWea['Description'] == 'broken clouds')  | (denWea['Description'] == 'fog') | (denWea['Description'] == 'haze')]
choices = ['3', '2']
denWea['DescRate'] = np.select(conditions, choices, default='0')
print(denWea)
```

```python
denWea['Celcius'].mean()
```

```python
conditions = [
    (denWea['Celcius'] <= 9),
    (denWea['Celcius'] >= 19)]
choices = ['0', '3']
denWea['TempRate'] = np.select(conditions, choices, default='2')
print(denWea)
```

```python
denWea.head()
```

```python
denWea['Wind Speed'].mean()
```

```python
conditions = [
    (denWea['Wind Speed'] <= 3),
    (denWea['Wind Speed'] > 3) | (denWea['Wind Speed'] < 6) ]
choices = ['3', '2']
denWea['WindRate'] = np.select(conditions, choices, default='0')
print(denWea)
```

```python
denWea.head()
```

```python
denWea.info()
```

```python
denWea['DescRate'] = denWea['DescRate'].astype(int)
```

```python
denWea['TempRate'] = denWea['TempRate'].astype(int)
```

```python
denWea['WindRate'] = denWea['WindRate'].astype(int)
```

```python
denWea.info()
```

```python
denWea['Overall'] = denWea["DescRate"] + denWea["TempRate"] + denWea["WindRate"]
```

```python
denWea.head()
```

```python
conditions = [
    (denWea['Overall'] >= 8), 
    (denWea['Overall'] >= 6) & (denWea['Overall'] < 8) ]
choices = ['Good', 'Moderate']
denWea['Weather'] = np.select(conditions, choices, default='Bad')
```
