### Identifying Wind and Losses

```python
conditions = [
    (nba['PTS.1'] > nba['PTS']),
    (nba['PTS'] == nba['PTS.1'])]
choices = ['Home Win', 'Draw']
nba['Result'] = np.select(conditions, choices, default='Away Win')
```

```python
nba.head()
```

```python
pt = nba.groupby('Visitor/Neutral')[('PTS')].count()
```

```python
pt
```

```python
nba['Difference'] = nba['PTS'] - nba['PTS.1']
```
