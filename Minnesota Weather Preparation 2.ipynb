```python
start_remove = pd.to_datetime('2014-12-31')
end_remove = pd.to_datetime('2012-01-01')
```

```python
wea = wea.loc[(wea.index < start_remove) & (wea.index > end_remove)]
```

```python
wea.head()
```

### Joining the teams

```python
nba = pd.read_csv("../../data/prep/700_NBA.csv", index_col="Date", parse_dates=True)
```

```python
nba.sort_index(ascending=True)
```

```python
bul = nba[(nba.values  == "Minnesota Timberwolves")]
```

```python
nfl = pd.read_csv("../../data/prep/800_NFL.csv", index_col="Date", parse_dates=True)
nfl.shape
```

```python
nfl.index.name = 'Date'
```

```python
nfl.head()
```

```python
nfl.sort_index(ascending=True)
```

```python
bea = nfl[(nfl.values  == "Vikings")]
```

```python
bea.head()
```

```python
fram1 = [bea, bul]
df = pd.concat(fram1)
```

```python
df.head()
```

```python
df = df.sort_index(ascending=True)
```

```python
conditions = [
    (df['Home/Neutral'] == 'Minnesota Timberwolves') | (df['Visitor/Neutral'] == 'Minnesota Timberwolves'),
    (df['Home/Neutral'] == 'Vikings') | (df['Visitor/Neutral'] == 'Vikings') ]
choices = ['NBA', 'NFL']
df['Game'] = np.select(conditions, choices, default='error')
```

```python
df
```

```python
df = df.drop(df.columns[[0, 1]], axis=1)
```

```python
df.head()
```

### Merging with the Weather

```python
comp = pd.merge(wea,df,how="left",right_index=True,left_index=True)
comp
```

```python
comp.index.duplicated()
```

```python
comp = comp[~comp.index.duplicated(keep='first')]
```

```python
comp
```

```python
comp['Game Day'] = comp.groupby("Weather")["Game"].ffill(limit=5)
```

```python
comp.head()
```

```python
comp['Game Day'] = comp.groupby("Weather")["Game Day"].bfill(limit=2)
```

```python
comp['Domestic'] = comp.groupby("Weather")["Game"].ffill(limit=7)
```

```python
comp
```
